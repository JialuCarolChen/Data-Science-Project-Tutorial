```python
# check is this a balanced classification problem
import collections
import matplotlib.pyplot as plt 
def distribution(data, column, title, label_rotation):
    y = data[column]
    #count the number of entities for each value in y
    histogram = collections.Counter(y)
    x = []
    y = []
    for key in histogram:
        x.append(key)
        y.append(histogram[key])
    #shift each bar to the left by 4; give each bar correct height; give each bar a width of 8
    plt.bar([i+1 for i, _ in enumerate(x)], y)
    # x-axis labels at 1, 2, 3, 4, 5
    plt.xticks([i+1 for i, _ in enumerate(x)], x, rotation=label_rotation)
    #plt.ylabel('number of responds')
    plt.title(title)
    plt.show()
    print(histogram)
distribution(marketing_Data, 'y', 'Has the client subscribed a term deposit?', 'horizontal')
distribution(marketing_Data, 'loan', 'Does client has loan?', 'horizontal')
distribution(marketing_Data, 'job', 'Types of jobs the client are doing', 'vertical')
distribution(marketing_Data, 'marital', 'Distribution of client marital status', 'horizontal')
distribution(marketing_Data, 'education', 'Distribution of client education level', 'horizontal')
distribution(marketing_Data, 'marital', 'Distribution of client marital status', 'horizontal')
distribution(marketing_Data, 'housing', 'Does the client have credit in defalut?', 'horizontal')
distribution(marketing_Data, 'contact', 'Distribution of contact communication type', 'horizontal')
distribution(marketing_Data, 'month', 'Distribution of last contact month of year ', 'horizontal')
distribution(marketing_Data, 'poutcome', 'Distribution of outcome of the previous marketing campaign', 'horizontal')
```

```python
# to balance the problem, delete rows that has value 'no' in column y, '0' in column previous, '-1' in column pdays
import random
delete_candidates_index = marketing_Data[(marketing_Data['y']=='no')].index.tolist()
number_of_candidates = len(delete_candidates_index)
number_of_delete = number_of_candidates - 5289
# sample a list of delete index (pick the rows randomly)
sample = random.sample(range(0,number_of_candidates), number_of_delete)
delete_index = []
for s in sample:
    delete_index.append(delete_candidates_index[s])
# delete the rows
marketing_Data = marketing_Data.drop(delete_index)
# reset the index
marketing_Data = marketing_Data.reset_index(drop=True)
#balance_marketing_Data
distribution(marketing_Data, 'y', 'Has the client subscribed a term deposit?', 'horizontal')
marketing_Data
```

```python
# for all the numeric attributes, draw box plots to find outliners and skewed data
import matplotlib.pyplot as plt
# get all the numeric data
num_vars = ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']
# function to make boxplot of all the numeric data
def make_boxplot(var):
    data = marketing_Data[var]
    plt.figure()
    plt.boxplot(data, 0, 'gD')
    plt.title(var)
    plt.show()
# make boxplot
for var in num_vars:
    make_boxplot(var)
```

```python
# calculate the not-outliners range of pdays and previous
def not_outliners_range(column):
    #calculate the IQR
    iq_range = marketing_Data[column].quantile(.75)-marketing_Data[column].quantile(.25)
    # the outliners has value greater than Q3+1.5*IQR or less than Q1-1.5*IQR.
    ut = marketing_Data[column].quantile(.75) + 1.5*iq_range
    lt = marketing_Data[column].quantile(.25) - 1.5*iq_range
    print("The outliners are outside of " + "(" + str(lt) + ", " + str(ut) + ")")
```

```python
# get the columns that has outliners, except pdays and previous
outliners_vars = ['age', 'balance', 'duration', 'campaign']
# build a function to delete outliners
def delete_outliners(column, pre_Data):
    #calculate the IQR
    iq_range = marketing_Data[column].quantile(.75)-marketing_Data[column].quantile(.25)
    # the outliners has value greater than Q3+1.5*IQR or less than Q1-1.5*IQR.
    ut = marketing_Data[column].quantile(.75) + 1.5*iq_range
    lt = marketing_Data[column].quantile(.25) - 1.5*iq_rang
    print("The outliners "+"of "+column+" are outside of " + "(" + str(lt) + ", " + str(ut) + ")")
    clean = pre_Data.loc[(pre_Data[column] <= ut)&(pre_Data[column] >= lt)]
    return clean

# delete the rows that has outliners, except pdays and previous
marketing_data_clean = marketing_Data
for var in outliners_vars:
    marketing_data_clean = delete_outliners(var, marketing_data_clean)
# take a look at the clean data set
marketing_data_clean.describe()
# reset the index
marketing_data_clean = marketing_data_clean.reset_index(drop=True)
marketing_data_clean.describe()
```

```python
distribution(marketing_data_clean, 'y', 'Has the client subscribed a term deposit?', 'horizontal')
distribution(marketing_data_clean, 'loan', 'Does client has personal loan?', 'horizontal')
distribution(marketing_data_clean, 'job', 'Types of jobs the client are doing', 'vertical')
distribution(marketing_data_clean, 'marital', 'Distribution of client marital status', 'horizontal')
distribution(marketing_data_clean, 'education', 'Distribution of client education level', 'horizontal')
distribution(marketing_data_clean, 'default', 'Has credit in default?', 'horizontal')
distribution(marketing_data_clean, 'housing', 'Has housing loan?', 'horizontal')
distribution(marketing_data_clean, 'contact', 'Distribution of contact communication type', 'horizontal')
distribution(marketing_data_clean, 'month', 'Distribution of last contact month of year ', 'horizontal')
distribution(marketing_data_clean, 'poutcome', 'Distribution of outcome of the previous marketing campaign', 'horizontal')
```

```python
#choose the input attributes for the classification problem 
distribution(marketing_data_clean, 'y', 'Has the client subscribed a term deposit?', 'horizontal')
marketing_data_clean.drop('previous', axis=1, inplace=True)
marketing_data_clean.drop('day', axis=1, inplace=True)
marketing_data_clean.drop('month', axis=1, inplace=True)
marketing_data_clean.drop('pdays', axis=1, inplace=True)
marketing_data_clean.drop('default', axis=1, inplace=True)
marketing_data_clean.drop('loan', axis=1, inplace=True)
marketing_data_clean.drop('campaign', axis=1, inplace=True)
marketing_data_clean
```

```python
#Categorical Encoding of the input variables
cols_to_transform = ['job','marital','education','housing','contact','poutcome'] 
categorical_encoded_data = pd.get_dummies(marketing_data_clean, columns = cols_to_transform)
#In order to pass the data into the scikit-learn decision tree we need to encode the target('y') to integers. 
df_mod = categorical_encoded_data.copy()
targets = df_mod['y'].unique()
map_to_int = {name: n for n, name in enumerate(targets)}
df_mod["Target"] = df_mod['y'].replace(map_to_int)
df_mod.drop('y', axis=1, inplace=True)
#df_mod it is the modified dataset for our classification
#'0' means no, '1' means yes, 
df_mod.columns
```

```python
# Note that scikit-learn expects each entry to be a list of values, e.g.,
# [ [val1, val2, val3],
#   [val1, val2, val3],
#   ... ]
# such that our input data set is represented as a list of lists
# We can extract the data in this format from pandas 
input_cols = ['age', 'balance', 'duration', 'job_admin.', 'job_blue-collar',
       'job_entrepreneur', 'job_housemaid', 'job_management', 'job_retired',
       'job_self-employed', 'job_services', 'job_student', 'job_technician',
       'job_unemployed', 'job_unknown', 'marital_divorced', 'marital_married',
       'marital_single', 'education_primary', 'education_secondary',
       'education_tertiary', 'education_unknown', 'housing_no', 'housing_yes',
       'contact_cellular', 'contact_telephone', 'contact_unknown',
       'poutcome_failure', 'poutcome_other', 'poutcome_success',
       'poutcome_unknown']
all_inputs = df_mod[input_cols].values
# Similarly, we can extract the target class
targetClass = df_mod['Target'].values
# Here's what a subset of our inputs looks like
all_inputs[:5]
```

```python
#split the data into traning and testing set
from sklearn.model_selection import train_test_split
(training_inputs,
 testing_inputs,
 training_classes,
 testing_classes) = train_test_split(all_inputs, targetClass, train_size=0.75, random_state=1)
```

```python
#Decision Tree classification
#There are several parameters that we can tune for decision tree classifiers
#but for now let's use a basic decision tree classifier
from sklearn.tree import DecisionTreeClassifier
# Create the classifier
decision_tree_classifier = DecisionTreeClassifier()
# Train the classifier on the training set
decision_tree_classifier.fit(training_inputs, training_classes)
# Validate the classifier on the testing set using classification accuracy
decision_tree_classifier.score(testing_inputs, testing_classes)
```

```python
#Parameter tuning
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
decision_tree_classifier = DecisionTreeClassifier()
parameters = {'max_depth': [2, 3, 4, 5, 6],
              'min_samples_split': [2, 3, 4, 5, 6],
              'criterion': ['entropy','gini'],
              'splitter': ['best','random']
             }
#Specify the splitting strategy of the data set (using random_state 1) 
sk = StratifiedKFold(n_splits=5, random_state=1)
grid_search = GridSearchCV(decision_tree_classifier,param_grid=parameters, cv=sk)
grid_search.fit(all_inputs, targetClass)
print('Best score: {}'.format(grid_search.best_score_))
print('Best parameters: {}'.format(grid_search.best_params_))
```

```python
# Use the best parameters to build the classifier
decision_tree_classifier = DecisionTreeClassifier(criterion='entropy', max_depth=6, min_samples_split=3, splitter='random')
# Train the classifier on the training set
decision_tree_classifier.fit(training_inputs, training_classes)
# Validate the classifier on the testing set using classification accuracy
decision_tree_classifier.score(testing_inputs, testing_classes)
```

```python
#Get the dummy classifier to simulate human random decision making
from sklearn.dummy import DummyClassifier
dc = DummyClassifier(strategy='stratified')
dc.fit(training_inputs, training_classes)
dc.score(testing_inputs, testing_classes)
```

```python
#Cross-validation on Decision Tree
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
#Specify the splitting strategy of the data set 
sk = StratifiedKFold(n_splits=20, random_state=1)
# to get a reasonable estimate of our classifier's performance
dt_f1_scores = cross_val_score(decision_tree_classifier, all_inputs, targetClass, cv=sk, scoring='f1')
print('The distribution of prediction accuracy score of the DecisionTree Classifer: '+ str(dt_f1_scores))
#The mean of the Decision Tree scores
print('The mean value of the distribution: '+str(sum(dt_f1_scores)/20))

#Cross-validation on tbe Dummy Classifier
dc_f1_scores = cross_val_score(dc, all_inputs, targetClass, cv=sk, scoring='f1')
print("The distribution of prediction accuracy score of the DummyClassifer: "+str(dc_f1_scores))
#The mean of the Decision Tree scores
print('The mean value of the distribution: '+str(sum(dc_f1_scores)/20))

#plotting the result
import numpy as np
x_axis = np.array(range(1,21))
plt.plot(x_axis, dt_f1_scores, 'ro')
plt.axis([0,21,0,1])
plt.ylabel('f1-measure Score')
plt.xlabel('Red: DecisionTreeClassifier  Blue: DummyClassifier')
plt.plot(range(1,21), dc_f1_scores, 'bs')
plt.show()
```

```python
import scipy.stats as stat
# Calculating the differences distribution u
u = dt_f1_scores - dc_f1_scores
print('The distribution of u is: '+str(u))
# calculate the standard deviation of u
sd = np.std(u)
print('Standard Deviation: '+str(sd))
# calculate the sample mean of u
mean = np.mean(u)
print('Sample mean: '+str(mean))
# calculate the t statistic
t = mean/(sd/np.sqrt(20))
print('t statistic: '+str(t))
# calculate the p-value P(t19>t)
p_Val = stat.t.sf(t, 19)
print('p-value P(t19>t): '+str(p_Val))
# compare p_Val with 0.05
print(str(p_Val-0.05))
```

```python
# Training the RandomForestClassifier to tun the best parameters
from sklearn.ensemble import RandomForestClassifier
random_forest_classifier = RandomForestClassifier()

parameter_grid = {'n_estimators': [5, 10, 25, 50],
                  'criterion': ['gini', 'entropy'],
                  'max_features': [1, 2, 3, 4],
                  'warm_start': [True, False]}

grid_search = GridSearchCV(random_forest_classifier,
                           param_grid=parameter_grid,
                           cv=sk)

grid_search.fit(all_inputs, targetClass)
print('Best score: {}'.format(grid_search.best_score_))
print('Best parameters: {}'.format(grid_search.best_params_))

grid_search.best_estimator_
```

```python
# use the best parameters
random_forest_classifier = grid_search.best_estimator_
# comparing decision tree with random forest
dt_scores = cross_val_score(decision_tree_classifier, all_inputs, targetClass, cv=sk, scoring='f1')
rf_scores = cross_val_score(random_forest_classifier, all_inputs, targetClass, cv=sk, scoring='f1')
#plotting the result
import numpy as np
x_axis = np.array(range(1,21))
plt.plot(x_axis, dt_scores, 'ro')
plt.axis([0,21,0,1])
plt.ylabel('f1-measure Score')
plt.xlabel('Red: DecisionTreeClassifier  Blue: RandomForestClassifier')
plt.plot(range(1,21), rf_scores, 'bs')
plt.show()
```
